```python
import pandas as pd
import spacy

with open('transfer_strings.txt', 'r') as file:
    strings = file.readlines()

# Load the NER model
nlp = spacy.load('en_core_web_sm')

# Initialize lists
senders = []
receivers = []
amounts = []
banks = []

# Process each string in the text file
for string in strings:
    doc = nlp(string)

    # Initialize variables
    sender = None
    receiver = None
    amount = None
    bank = None

    # Extract named entities
    for entity in doc.ents:
        if entity.label_ == 'PERSON' and not sender:
            sender = entity.text.strip()
        elif entity.label_ == 'PERSON' and not receiver:
            receiver = entity.text.strip()
        elif entity.label_ == 'MONEY':
            amount = entity.text.strip()
        elif entity.label_ == 'ORG':
            bank = entity.text.strip()

    # Append extracted information to respective lists
    senders.append(sender)
    receivers.append(receiver)
    amounts.append(amount)
    banks.append(bank)

# Create a dataframe
data = {
    'Sender': senders,
    'Receiver': receivers,
    'Amount Transferred': amounts,
    'Bank Name': banks
}
df = pd.DataFrame(data)

# Print the dataframe
display(df)
```
